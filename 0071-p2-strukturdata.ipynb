```python
# Tuple
my_tuple = (1, "Hello", 3.14, True)

# Mengakses elemen Tuple
print("Elemen pertama:", my_tuple[0])
print("Elemen kedua:", my_tuple[1])
print("Elemen ketiga:", my_tuple[2])
print("Elemen keempat:", my_tuple[3])
```

```python
# Dictionary
my_dict = {"Apple": 10, "Banana": 5, "Orange": 8}

# Mengakses nilai berdasarkan kunci
print("Jumlah apel:", my_dict["Apple"])
print("Jumlah pisang:", my_dict["Banana"])
print("Jumlah jeruk:", my_dict["Orange"])

# Iterasi melalui Dictionary
print("Isi Dictionary:")
for key, value in my_dict.items():
    print(f"{key}: {value}")
```
